```python
import matplotlib.pyplot as plt

# Sample data
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
avg_temperatures = [30, 31, 35, 42, 39]

# Plotting the line graph for temperature trends
plt.figure(figsize=(8, 6))
plt.plot(months, avg_temperatures, marker='o', color='red', linestyle='-', linewidth=2, markersize=8)

# Adding labels and title
plt.xlabel('Months')
plt.ylabel('Average Temperature (Â°C)')
plt.title('Temperature Trends')

# Displaying the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

year_sem = ['1stYear_1stSem', '1stYear_2ndSem', '1stSem_Summer', '2ndYear_1stSem', '2ndYear_2ndSem', '2ndYear_Summer']
gwa = [1.39, 1.34, 1.42, 1.33, 1.47, 1.08]

plt.figure(figsize=(12, 6))
plt.plot(year_sem, gwa, marker='o', color='green', linestyle='-', linewidth=2, markersize=8)

plt.xlabel('Year and Semester')
plt.ylabel('GWA')
plt.title('GWA Records')

plt.gca().invert_yaxis()

plt.show()
```

```python
import matplotlib.pyplot as plt

desserts = ['Halo-halo', 'Milkshake', 'Ice cream', 'Cake', 'Fruit Salad', 'Milktea']
sales = [102, 119, 390, 295, 45, 303]

plt.figure(figsize=(8, 6))
plt.bar(desserts, sales, color='pink')

font1 = {'family':'serif','color':'magenta','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title('Most Sold Dessert Product', fontdict = font2)
plt.xlabel('Top Desserts', fontdict = font1)
plt.ylabel('Sales', fontdict = font1)


plt.show()
```
